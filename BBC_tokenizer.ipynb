```python
# Buat Stop words
def remove_stopwords(sentence):
  stopwords = ['you' , 'we', 'they', 'are']

  sentence = sentence.lower()

  words = sentence.split()
  result_sentence = [word for word in words if word not in stopwords]
  sentence = ''.join(result_sentence)

  return(sentence)
```

```python
# Membaca data mentah
def parse_data_from_file(filename):
  sentence = []
  labels = []

  df = pd.read_csv(filename)

  labels = df['category'].apply(remove_stopwords).tolist()
  sentence = df['text'].apply(remove_stopwords).tolist()

  return sentence, labels
```

```python
sentence, labels = parse_data_from_file('dataset/bbc-text.csv')
print('ORIGINAL DATASET')
print('sentence have : ', len(sentence))
print('sentence after remove stopwords: ', len(sentence[0].split()))
print('first sentence : ', sentence[1])
print('sentence remove stopwords: ', sentence[0].split())
print('labels have: ', len(labels))
```

```python
# Buat fungsi Tokenizer
def fit_tokenizer(sentence):
  tokenizer = Tokenizer(oov_token = 'OOV')
  tokenizer.fit_on_texts(sentence)
  return tokenizer
```

```python
tokenizer = fit_tokenizer(sentence)
wordIndex = tokenizer.word_index
print(wordIndex)
```

```python
# Buat fungsi Padded
def get_padded_sequences(tokenizer, sentence):
  sequences = tokenizer.texts_to_sequences(sentence)
  padded = pad_sequences(sequences, padding='post')

  return padded
```

```python
pad = get_padded_sequences(tokenizer, sentence)
print(pad[1])
print(pad.shape)
```
